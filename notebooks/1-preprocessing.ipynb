###Notebook 1: Preprocessing - Exploratory Data Analysis (EDA)###

Here, I conducted an exploratory data analysis on the Singapore Airlines Reviews dataset. After loading the data into a pandas DataFrame, following was done:

Summarized the Dataset: I used descriptive statistics to understand the central tendency, dispersion, and shape of the dataset's distribution. Before now manual data cleaning was done to check for missing cells etc. 
Plotted the Distribution of Ratings: visualized the distribution of customer ratings to understand the balance of classes in the dataset.
Visualized the Length of Reviews: analyzed the length of the review texts as longer or shorter reviews may carry different weights in sentiment.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading dataset
reviews_df = pd.read_csv('/Users/adese/Downloads/Singapore_Airlines_Reviews New.csv')
reviews_df.head()

# column 'A' is 'rating' and column 'B' is 'text'
reviews_df.columns = ['rating', 'text']

# 1. Summarized the Dataset
print(reviews_df.describe())  
print(reviews_df.info())      

# 2. Checked for Null Values
print(reviews_df.isnull().sum())

# 3. Plotting: Distribution of Ratings
sns.histplot(reviews_df['rating'], kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# 4. Visualizing: Length of Reviews
reviews_df['text_length'] = reviews_df['text'].apply(lambda x: len(str(x)))
sns.histplot(reviews_df['text_length'], kde=True)
plt.title('Distribution of Review Text Lengths')
plt.xlabel('Review Text Length')
plt.ylabel('Frequency')
plt.show()

```
